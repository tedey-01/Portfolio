```python
data=[]
with open("hits.txt") as f:
    for line in f:
        data.append([x for x in line.split()])
```

```python
data
```

```python
ip = []
for element in data:
    ip.append(element[1])  
ip
```

```python
from collections import Counter
bester= Counter(ip)
for el in bester.most_common(5):
    print('{}'.format(el[0]))
```
