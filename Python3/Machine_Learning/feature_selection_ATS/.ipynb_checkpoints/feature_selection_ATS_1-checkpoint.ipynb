### Посмотрим на распределение целевого класса – оттока клиентов.

```python
df["Churn"].value_counts()
```

```python
df["Churn"].value_counts().plot(kind= "bar", label= 'Churn')
plt.legend()
plt.title('Распределение оттока клиентов');
```

##### По корреляционной матрице посмотрим какие из количественных признаков имеют линейную зависимость 

```python
#Уберём все котегориальные и бинарные признаки 
matr_of_corr = df.drop(['State', 'International plan', 'Voice mail plan', 'Area code'], axis = 1).corr()
```

```python
sns.heatmap(matr_of_corr)
```

```python
features = list(set(df.columns) - set(['State', 'International plan', 'Voice mail plan',  'Area code',
                                      'Total day charge',   'Total eve charge',   'Total night charge',
                                        'Total intl charge', 'Churn']))
print(features)
```

```python
df[features].hist(figsize=(20,12));
```

### Дальше посмотрим, как признаки связаны с целевым – с оттоком.

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(18, 12))

for idx, feat in  enumerate(features):
    sns.boxplot(x='Churn', y=feat, data=df, ax=axes[int(idx / 4), int(idx % 4)])
```

```python
sns.boxplot(x='Churn', y='Total day minutes', data=df);
```

### Теперь изобразим распределение числа обращений в сервисный центр

```python
sns.countplot(x='Customer service calls', hue='Churn', data=df);
```

### Проделаем то же самое для  бинарных признаков International plan и Voice mail plan с оттоком.

```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))

sns.countplot(x='International plan', hue='Churn', data=df, ax=axes[0]);
sns.countplot(x='Voice mail plan', hue='Churn', data=df, ax=axes[1]);
```

### Объединим рассмотренные выше условия и построим сводную табличку для этого объединения и оттока.

```python
df['Many service calls'] = (df['Customer service calls'] > 3).astype('int')

pd.crosstab(df['Many service calls'] & (df['International plan']=="Yes")  , df['Churn'])
```
